```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
temperature = np.array([40, 27, 40, 73, 64, 34, 9, 8, 23, 63, 65, 41, 21, 38, 58])
insulation = np.array([3, 3, 10, 6, 6, 6, 6, 10, 10, 3, 10, 6, 3, 3, 10])
oil_used = np.array([275.3, 363.8, 164.3, 40.8, 94.3, 230.9, 366.7, 300.6, 237.8, 121.4,
                     31.4, 203.5, 441.1, 323.0, 52.5])

# Combine features into a 2D array
features = np.column_stack((temperature, insulation))

# Create a linear regression model
model = LinearRegression()

# Fit the model
model.fit(features, oil_used)

# Make a prediction
prediction = model.predict([[15, 10]])

print("Predicted heating oil consumption (gallons):", prediction[0])
```

```python
# Question Two
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Put the data
length = [3, 4, 5, 6, 7]
made = [84, 88, 61, 61, 44]
missed = [17, 31, 47, 64, 90]

# Step 2: Create input (X) and output (y) data
X = []
y = []

for i in range(len(length)):
    X += [length[i]] * (made[i] + missed[i])  # Repeat length for each putt
    y += [1] * made[i] + [0] * missed[i]       # 1 for made, 0 for missed

# Step 3: Convert to numpy arrays and reshape
X = np.array(X).reshape(-1, 1)
y = np.array(y)

# Step 4: Train the logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Step 5: Make predictions
test_lengths = np.linspace(2, 8, 100).reshape(-1, 1)
predicted_probs = model.predict_proba(test_lengths)[:, 1]

# Step 6: Plot the results
plt.plot(test_lengths, predicted_probs, label='Predicted Probability')
plt.scatter(length, np.array(made) / (np.array(made) + np.array(missed)), color='red', label='Actual Proportion')
plt.xlabel("Putt Length (feet)")
plt.ylabel("Probability of Making Putt")
plt.title("Logistic Regression for Golf Putts")
plt.legend()
plt.grid(True)
plt.show()
```
